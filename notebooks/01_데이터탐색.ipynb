```python
import pandas as pd
```

```python
# encoding을 cp949로 변경
store_info = pd.read_csv('../data/raw/store_info.csv', encoding='cp949')
estimated_sales = pd.read_csv('../data/raw/estimated_sales.csv', encoding='cp949')
closure_rate = pd.read_csv('../data/raw/closure_rate.csv', encoding='cp949')
rent_price = pd.read_csv('../data/raw/rent_price.csv', encoding='cp949')
resident_population = pd.read_csv('../data/raw/resident_population.csv', encoding='cp949')
working_population = pd.read_csv('../data/raw/working_population.csv', encoding='cp949')
subway_passengers = pd.read_csv('../data/raw/subway_passengers.csv', encoding='utf-8')
```

```python
dfs = {
    'store_info': store_info,
    'estimated_sales': estimated_sales,
    'closure_rate': closure_rate,
    'rent_price': rent_price,
    'resident_population': resident_population,
    'working_population': working_population,
    'subway_passengers': subway_passengers,
}

for name, df in dfs.items():
    print(f'\n========== {name} ==========')
    print(list(df.columns))
```
